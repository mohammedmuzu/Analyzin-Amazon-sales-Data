```python
import pandas as pd

# Load the dataset
amazon_data = pd.read_csv('/content/drive/MyDrive/Amazon sales data /Amazon Sales data.csv')

# Extract relevant columns
sales_data = amazon_data[['Order Date', 'Total Revenue']]

```

```python
# Convert 'Order Date' column to datetime format
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])

# Extract month and year
sales_data['Month'] = sales_data['Order Date'].dt.month
sales_data['Year'] = sales_data['Order Date'].dt.year

# Group by Year and Month and sum the total revenue
year_monthly_sales = sales_data.groupby([sales_data['Year'], sales_data['Month']])['Total Revenue'].sum()

# Group by Year and sum the total revenue
yearly_sales = sales_data.groupby('Year')['Total Revenue'].sum()
```

# Data Visualizations

# **Sales Trend**

Monthly Sales Trend

```python
import matplotlib.pyplot as plt

# Plotting sales trends
year_monthly_sales.plot(kind='line', figsize=(10, 6), title='Monthly Sales Trend')
plt.xlabel('Year, Month')
plt.ylabel('Total Revenue ($)')
plt.grid(True)
plt.show()
```

Yearly Sales Trend

```python
yearly_sales.plot(kind='bar', figsize=(10, 6), title='Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Revenue ($)')
plt.grid(True)
plt.show()
```

# **Key Metrics and Factors**

Region-wise Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt


region_sales = amazon_data.groupby('Region')['Total Revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
region_sales.plot(kind='bar', color='skyblue')
plt.title('Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Item Type Analysis

```python
item_sales = amazon_data.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
item_sales.plot(kind='bar', color='lightgreen')
plt.title('Total Profit by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Profit ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

 Sales Channel Analysis

```python
channel_sales = amazon_data.groupby('Sales Channel')['Total Revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 6))
channel_sales.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightblue', 'lightgreen'])
plt.title('Total Revenue by Sales Channel')
plt.ylabel('')
plt.show()

```

Order Priority Analysis

```python

priority_sales = amazon_data.groupby('Order Priority')['Total Revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 6))
priority_sales.plot(kind='bar', color='orange')
plt.title('Total Revenue by Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

```

Unit Price Analysis

```python

plt.figure(figsize=(10, 6))
plt.scatter(amazon_data['Unit Price'], amazon_data['Total Revenue'], color='blue', alpha=0.5)
plt.title('Total Revenue vs Unit Price')
plt.xlabel('Unit Price ($)')
plt.ylabel('Total Revenue ($)')
plt.grid(True)
plt.show()

```

Unit Cost Analysis

```python
plt.figure(figsize=(10, 6))
plt.scatter(amazon_data['Unit Cost'], amazon_data['Total Revenue'], color='green', alpha=0.5)
plt.title('Total Revenue vs Unit Cost')
plt.xlabel('Unit Cost ($)')
plt.ylabel('Total Revenue ($)')
plt.grid(True)
plt.show()
```

Time-series Forecasting (Example using matplotlib)

```python

plt.figure(figsize=(10, 6))
plt.plot(amazon_data['Order Date'], amazon_data['Total Revenue'], color='purple')
plt.title('Total Revenue Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Revenue ($)')
plt.grid(True)
plt.show()
```
